```python
grouped_gender = purchase_data.groupby('Gender')['SN'].nunique()
avg_gender = (grouped_gender / 576.00) * 100.00
#grouped_gender

avg_gender
```

```python
gender2 =purchase_data.groupby('Gender')['Price']
pc = gender2.count()
app = gender2.mean()
tpv = gender2.sum()
atpp = tpv / grouped_gender
atpp

```

```python
gender_pur_df = pd.DataFrame({
    "Gender":grouped_gender,
    "Purchase Count":pc,
    "Average Purchase Price":app,
    "Total Purchase Value": tpv,
    "Avg Total Purchase per Person": atpp
})
gender_pur_df
```

```python
  
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]



```

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
purchase_data.head()
```

```python
age_df = purchase_data.groupby('Age Ranges')
age_df['Age Ranges'].count()
```

```python
gpc = age_df["Price"].count()
gapp = age_df["Price"].mean()
gtpv = age_df["Price"].sum()
buyers_by_age= age_df['SN'].nunique()
ave_per_buyer = gtpv/buyers_by_age


```

```python
age_df = pd.DataFrame({
   "Purchase Count":gpc,
    "Average Purchase Price":gapp,
    "Total Purchase Value":gtpv,
    "Avg Total Purchase per Person":ave_per_buyer
})
age_df
```

```python
user_group = purchase_data.groupby('SN')

#user_group.sort_values(by=['SN'])
num_pur_by_user = user_group['SN'].count()
user_app = user_group['Price'].mean()
user_tpp  = user_group['Price'].sum()

```

```python
top_user_df = pd.DataFrame({
    "Purchase Count": num_pur_by_user ,
    "Average Purchase Price":user_app ,
    "Total Purchase Value":user_tpp
})
top_user_df.head(90)    
```

```python
top_user_sorted_df = top_user_df.sort_values("Purchase Count", ascending=False)
top_user_sorted_df.head()
```
